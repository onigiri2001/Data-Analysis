```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager

# ----------- 初始化 Chrome -----------
options = Options()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
wait = WebDriverWait(driver, 30)

# ----------- 開啟 LEBRON 頁面 -----------
driver.get("https://bball-index.shinyapps.io/Lebron/")
time.sleep(8)

# ----------- 設定 Metrics -----------
metrics = ["LEBRON WAR", "LEBRON", "O-LEBRON", "D-LEBRON"]
driver.execute_script(f"$('#Metric')[0].selectize.setValue({metrics});")

# ----------- 設定 Minutes 滑桿範圍 -----------
driver.execute_script("""
  $('#Minutes').data("ionRangeSlider").update({ from: 1, to: 3240 });
""")

# ----------- 按年份逐一爬取資料 -----------
for year in range(2014, 2026):
    print(f"📅 開始擷取 {year} 年資料...")
    
    # 更新年份滑桿
    driver.execute_script(f"""
        $('#Years').data("ionRangeSlider").update({{ from: {year}, to: {year} }});
    """)
    time.sleep(1)

    # 點擊 Run Query
    run_btn = wait.until(EC.element_to_be_clickable((By.ID, "make_table")))
    run_btn.click()
    print("▶️ 已點擊 Run Query")
    
    # 等資料表格出現
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "table")))
    time.sleep(2)

    # 每頁顯示 100 筆
    driver.execute_script("""$('select[name$="_length"]').val(100).trigger('change');""")
    time.sleep(2)

    # 分頁擷取資料
    all_data = []
    page = 1
    while True:
        try:
            table = wait.until(EC.presence_of_element_located((By.TAG_NAME, "table")))
            html = table.get_attribute("outerHTML")
            df = pd.read_html(html)[0]
            all_data.append(df)
            print(f"✅ 第 {page} 頁擷取 {len(df)} 筆")
            page += 1

            # 檢查是否有下一頁
            next_btn = driver.find_element(By.CSS_SELECTOR, 'a.paginate_button.next')
            if "disabled" in next_btn.get_attribute("class"):
                print("⛔ 沒有更多頁了")
                break
            else:
                next_btn.click()
                time.sleep(2)
        except Exception as e:
            print("⚠️ 發生錯誤，跳出迴圈：", e)
            break

    # 合併並存檔
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        final_df.to_csv(f"lebron_{year}.csv", index=False, encoding="utf-8-sig")
        print(f"💾 儲存完成：lebron_{year}.csv（{len(final_df)} 筆）")
    else:
        print(f"⚠️ {year} 無法擷取或無資料")

    print("-" * 50)
    time.sleep(3)

```
